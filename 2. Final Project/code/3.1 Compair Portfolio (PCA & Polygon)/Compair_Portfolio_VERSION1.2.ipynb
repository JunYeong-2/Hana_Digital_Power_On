```python
import numpy as np
import pandas as pd
import seaborn as sns
import math
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
```

# Load Data

```python
df = pd.read_csv('/content/drive/MyDrive/hana/data/final_data.csv', encoding='cp949')
```

# Example.

```python
def checkpoint(target_name):
  # 초기값 설정
  target = df.loc[df['종목명'] == target_name]
  df_del_samsung = df.loc[df['종목명'] != target_name]

  # 딕셔너리 초기화
  distance_dict = {}

  # 각 종목에 대해 거리 계산 및 딕셔너리에 추가
  for stock in df_del_samsung.values:
      distance = math.sqrt((stock[1] - target['PC1'])**2 + (stock[2] - target['PC2'])**2)
      distance_dict[stock[0]] = distance

  # 거리를 기준으로 딕셔너리 정렬
  sorted_distance = sorted(distance_dict.items(), key=lambda x: x[1])

  # 결과 출력
  closest_stock, min_distance = sorted_distance[0]
  print(f"{target_name}와 가장 가까운 종목은 {closest_stock}이며, 거리는 {min_distance}입니다.")

  print(sorted_distance[:3])
```

```python
checkpoint('삼성전자')
```

```python
checkpoint('하나금융지주')
```
